```python
salary_list = [[j.replace(" ", "") for j in i.split('\t')] for i in postdoc_salary_string.split('\n')]
salary_df = pd.DataFrame(salary_list[2:-1])
salary_df.columns = ["year", 'PREDOC', "0", "1", "2", "3", "4", "5", "6", "7"]
salary_df = salary_df.apply(lambda x: x.str.strip()).replace('', np.nan)
for col in salary_df.columns.values:
    salary_df[col] = salary_df[col].astype(float)
salary_df 
```

```python
import cpi
```

```python
from datetime import date
```

```python
levels = ['PREDOC', "0", "1", "2", "3", "4", "5", "6", "7"]
```

```python
inflated_salary_df = pd.DataFrame(columns = ["year"] + levels)
for idx, row in salary_df.iterrows():
    inflated_salary_df.loc[row.year, "year"] = inflated_value = row.year
    for level in levels:
        try:
            inflated_value = cpi.inflate(
                        int(row[level]),
                        year_or_month=date(int(row.year), 1, 1),
                        to=date(2021, 1, 1),
                    )
        except:
            inflated_value = np.nan
        inflated_salary_df.loc[row.year, level] = inflated_value
```
